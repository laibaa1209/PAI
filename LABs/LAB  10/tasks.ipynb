```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("dataset.csv")

plt.figure(figsize=(12, 12))
sns.heatmap(data=df.corr(), annot=True)
plt.show()
```

```python
count = df['sex'].value_counts()
plt.pie(count, autopct="%1.1f%%",labels=['male', 'female'], colors=["purple","yellow"])
plt.show()
```

```python
sns.displot(df['chol'], edgecolor=None, kde=True)
plt.show()
```

```python
sns.displot(df['thalach'],  edgecolor=None, kde=True)
plt.show()
```
